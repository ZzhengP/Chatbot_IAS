```python
import os, re
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from collections import Counter
import pandas as pd
from mpl_toolkits import mplot3d
```

```python
data=pd.read_csv('data.tsv',sep='\t')
dataQ=data.Question
Q=[]
Q2W=[]# liste de mots pour chaque phrase 
for i in range(len(dataQ)):
    Q.append(dataQ[i])
for each in Q:
    Q2W.append(each.split(" "))
#print(Q2W)
```

```python
#data.describe()
```

```python
import gensim, logging
```

```python
window_size=2
N=len(Q2W)
vector_size=3
model=gensim.models.Word2Vec(Q2W,vector_size=vector_size, window=2, min_count=1, workers=4)
```

```python
V=[]
for i in range (N):
    s=np.zeros(vector_size)
    for j in range(len(Q2W[i])):
        wv=np.array(model.wv.get_vector(Q2W[i][j]))
        s+=wv
    V.append(list(s))
print(V[1])
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
for i in range(100):
    #ax.scatter3D(V[i][0],V[i][1],V[i][2],label=Q2W[i])
    ax.scatter3D(V[i][0],V[i][1],V[i][2])
plt.legend()
```
